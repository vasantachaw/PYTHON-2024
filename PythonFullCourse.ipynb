```python
class A:
    def method(self):
        print("Method from A")


class B(A):
    def method(self):
        print("Method from B")
        super().method()


class C(B):
    def method(self):
        print("Method from C")
        super().method()


class D(C):
    def method(self):
        print("Method from D")
        super().method()


d = D()
d.method()
# Output:
# Method from D
# Method from B
# Method from C
# Method from A
```

```python
class A:
    def method(self):
        print("Method from A")


class B(A):
    def method(self):
        print("Method from B")
        super().method()


class C(A):
    def method(self):
        print("Method from C")
        super().method()


class D(B, C):
    def method(self):
        print("Method from D")
        super().method()


d = D()
d.method()
# Output:
# Method from D
# Method from B
# Method from C
# Method from A
```

```python
from tqdm import tqdm
import time


class SBI_ATM:
    balance = 0
    pin = 0

    def __init__(self):
        print(" WE L C O M E   T O    S B I    A T M \n")
        print(" Please Insert Your Card .")
        for i in tqdm(range(101), desc=" Loading", ascii=False, ncols=75):
            time.sleep(0.05)
        print("\n 1. Set Your Pin Code .")
        print(" 2. Deposite Your Amount .")
        print(" 3. Withdraw Your Amount .")
        print(" 4. Inqury Your blance .")
        print(" 5. Cancel .\n")
       
    def enter_pin(self):
        self.pin_code = int(input(" Enter Your Pin Code : "))

    def enter_amount(self):
        self.amount = int(input(" Enter Your amount : "))

    def set_your_pin(self):
        self.pin =int(input(' Set Your New Pin Code :'))
        
        print(" Successfully Set your Pin !")

    def deposite_amount(self):
        try:
            self.enter_pin()
            if self.pin_code == self.pin:
                self.enter_amount()
                self.balance += self.amount
                print(" Successfully Deposited !")
            else:
                print(' Invalid Pin Code !')
        except ValueError as e:
            print(e)

    def inqury_balannc(self):
        print(f" Your Total Amount : {self.balance}")

    def withdraw_amount(self):
        try:
            self.enter_pin()
            if self.pin == self.pin_code:
                self.enter_amount()
                if self.balance > self.amount and self.balance > 100:
                    self.balance -= self.amount
                    print(f" Successfully Withdraw : {self.amount}")
                else:
                    print(" Insufficiant balance !")
            else:
                print(' Invalid Pin Code !')
        except ValueError as e:
            print(e)


if __name__ == "__main__":
    sbi = SBI_ATM()
    while True:
        chose=int(input())
        match chose:
            case 1:
                sbi.set_your_pin()

            case 2:
                sbi.deposite_amount()

            case 3:
                sbi.withdraw_amount()

            case 4:
                sbi.inqury_balannc()            
            case 5:
                break

          
```
